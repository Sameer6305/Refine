> `candidates.jsonl` is gitignored. Place it at `d:/Refine/candidates.jsonl` before running.

```python
# ── Dependencies ──────────────────────────────────────────────────────────── #
import json
import pathlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

CANDIDATES_PATH = pathlib.Path("../candidates.jsonl")
assert CANDIDATES_PATH.exists(), f"File not found: {CANDIDATES_PATH.resolve()}"
```

```python
# ── 1. Load & schema inspection ───────────────────────────────────────────── #
records = []
with CANDIDATES_PATH.open() as f:
    for line in f:
        line = line.strip()
        if line:
            records.append(json.loads(line))

df = pd.DataFrame(records)
print(f"Total candidates: {len(df):,}")
df.dtypes
```

```python
# ── 2. Field completeness ─────────────────────────────────────────────────── #
missing = df.isnull().mean().sort_values(ascending=False)
missing[missing > 0].plot(kind="bar", figsize=(12, 4), title="Missing value rate by field")
plt.tight_layout()
plt.show()
```
